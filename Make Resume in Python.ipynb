```python
def main():
    # Take input from user
    name = input("Enter your name: ").upper()
    age = int(input("Enter your age: "))
    siblings_ages = []
    for i in range(int(input("Enter the number of siblings you have: "))):
        sibling_name = input(f"Enter sibling {i+1} name: ").capitalize()
        sibling_age = int(input(f"Enter {sibling_name}'s age: "))
        siblings_ages.append((sibling_name, sibling_age))
    income = float(input("Enter your income: "))
    classes = input("Enter your class details: ")
    hobbies = get_hobbies()
    father_business = input("Enter your father's business: ")
    business_address = input("Enter business address: ")
    siblings_school_address = {}
    for sibling_name, _ in siblings_ages:
        siblings_school_address[sibling_name] = input(f"Enter {sibling_name}'s school address: ")
    
    # Store values in dictionaries
    data = {
        "Name": name,
        "Age": age,
        "Siblings": dict(sorted(siblings_ages)),
        "Income": income,
        "Classes": classes,
        "Hobbies": hobbies,
        "Father Business": father_business,
        "Business Address": business_address,
        "Siblings School Address": siblings_school_address
    }

    # Print the dictionary
    print(data)

    # Save data to a file
    save_to_file(data)


def get_hobbies():
    hobbies = []
    while True:
        hobby = input("Enter a hobby (or 'done' to finish entering hobbies): ")
        if hobby.lower() == 'done':
            break
        hobbies.append(hobby)
    return hobbies


def save_to_file(data):
    with open("fakir_resume.txt", "w") as file:
        for key, value in data.items():
            file.write(f"{key}: {value}\n")


if __name__ == "__main__":
    main()
```
